# Effect of cross-validation on distance estimators

```python
import seaborn as sb
import matplotlib.pyplot as plt
from sim_crossval import crossval_sim  
```

```python
def plot_panel(T):
    sb.violinplot(data=T,x='True',y='Simp',inner='quart',color='red')
    sb.violinplot(data=T,x='True',y='Cross',inner='quart',color=[0.1,0.25,0.95])
    sb.despine()
    plt.plot([0,3],[0,3],'k--')
    plt.xlabel('True distance')
    plt.ylabel('Estimated distance')
    ax=plt.gca()
    ax.set_ylim([-2,9])


```

```python
T1 = crossval_sim(sigma='iid',n_sim=100)
T2 = crossval_sim(sigma='neigh',n_sim=100)
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_panel(T1)
plt.subplot(1,2,2)
plot_panel(T2)
plt.savefig('crossval_example.pdf')
```
